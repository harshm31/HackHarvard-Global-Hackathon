```python
import pandas as pd 
import numpy as np 
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense , Dropout , BatchNormalization
from tensorflow.keras.callbacks import TensorBoard , ModelCheckpoint
import time 
from sklearn import preprocessing
```

```python
df = pd.read_csv("clean_weather.csv")
df.columns
df.drop(columns=["Unnamed: 0"],inplace=True)

df.head(5)
```

```python
temp_df = df["Mean Temp (C)"]
#temp_df.head(10)

df.drop(columns=["Mean Temp (C)","Year","Month","Day"],inplace=True)
```

```python
#df.head(20)
```

```python
X = []
Y = []

#normalize columns in df 
cols_to_norm = ["Total Rain (mm)","Total Snow (mm)","Total Precip (mm)"]
df[cols_to_norm] = df[cols_to_norm].apply(lambda x : (x-x.min()) / (x.max()-x.min()))

df.head(5)
```

```python
for i in df.values :
    X.append(i)
    
X = np.array(X)

for i in temp_df.values :
    Y.append(i)

Y = np.array(Y)
print(X[1])
print(Y[1])
```

```python
len(X)
```

```python
train_x = X[:2600]
train_y = Y[:2600]

test_x = X[2600:]
test_y = Y[2600:]
```

```python
#for 1st run 
print(train_x.shape)
epochs = 100
batch_size = [16]

name = "1st_model"
tensorboard = TensorBoard(log_dir="manylogs/{}".format(name))

model = Sequential()

model.add(Dense(7,input_dim = 7,activation="relu"))
#model.add(Dropout(0.2))

model.add(Dense(3, activation="relu"))

#model.add(Dense(3,activation="relu"))
#model.add(Dropout(0.2))

#model.add(Dense(128,activation="relu"))

model.add(Dense(1,activation="relu"))

model.compile(loss="mean_squared_error",optimizer="adam",metrics=["accuracy"])
```

```python
#fit model
for b in batch_size :
    print("\n batch size : {} \n".format(b))
    model.fit(train_x,train_y,epochs=epochs,batch_size=b,callbacks=[tensorboard],validation_split=0.0)
```

```python
for i in range(5):
    print(test_y[i])
```

```python
preds = model.predict([test_x])
print(preds[1:6])
```
